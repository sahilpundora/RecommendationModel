# Part A - Basic Understanding of Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

## interests.csv

```python
df = pd.read_csv("interests.csv") 

#strip the leading or trailing whitespace
df.columns = ['user_id', 'category']
#df = df.rename(columns={' category ': 'category'})

df.head()

users = len(df['user_id'].unique())
print('How many users are there? ', users)

interests = len(df['category'].unique())
print('How many interest categories were created by users? ', interests)
```

## New dataframe grouped by User ID with count of unique interests

```python
df_sort = df.groupby(by='user_id', as_index=False).agg({'category': pd.Series.nunique})
df_sort.columns = ['user_id', '#_category']
print(df_sort)
```

## Overview Plot

```python
df_sort['#_category'].plot()
plt.title('Number of declared interests by each user', color='black')
plt.xlabel('Users')
plt.ylabel('Number of declared interests')
plt.show()
```

## Histogram - declared interests

```python
df_sort['#_category'].hist(bins=100)
plt.title('Histogram of number of declared interests', color='black')
plt.xlabel('Number of declared interests')
plt.ylabel('Frequency - Number of users')
plt.show()
```

## Boxplot - declared interests

```python
df_sort['#_category'].plot(kind='box')
plt.title('Histogram of number of declared interests', color='black')
plt.show()
print(df_sort['#_category'].describe())
```
